```python
df_model.dropna(inplace=True)

# создаем модель и обучаем ее
X = sm.add_constant(df_model[['ВВП ( ценах 2021 г., млрд. руб.)','Уровень безработицы населения в возрасте 15-72 лет','произошедшие изменения личного материального положения за год','благоприятность условий для крупных покупок','благоприятность условий для сбережений','Среднедушевые денежные доходы населения ( руб.)','Ср.знач ключевая ставки за квартал, % годовых','Ср. знач. инфляция за квартал, % г/г']])
y = df_model['Количество договоров микрозайма (всего) за за отчетный квартал), ед.']
model = sm.OLS(y, X).fit()

# выводим результаты
print(model.summary())
```

Эта модель является линейной регрессионной моделью, которая пытается предсказать количество договоров микрозайма (всего) за отчетный квартал на основе десяти независимых переменных: ВВП, уровня безработицы населения, индекса уверенности, индекса текущего личного материального положения, произошедшие изменения личного материального положения за год, благоприятность условий для крупных покупок, благоприятность условий для сбережений, среднедушевых денежных доходов населения, среднего значения ключевой ставки за квартал и среднего значения инфляции за квартал.

Коэффициенты перед каждой независимой переменной указывают, насколько сильно она влияет на зависимую переменную. Например, коэффициент перед переменной ВВП отрицательный (-212.5270), что означает, что увеличение ВВП связано с уменьшением количества договоров микрозайма.

R-квадрат (R-squared) показывает, как хорошо модель подходит для данных. В данном случае R-квадрат равен 0,968, что означает, что 96,8% дисперсии зависимой переменной (количество договоров микрозайма) может быть объяснено независимыми переменными.

Adj. R-квадрат является скорректированным R-квадратом и принимает в расчет количество независимых переменных, которые были использованы в модели.

F-статистика используется для проверки значимости модели. В данном случае p-значение равно 0,00390, что означает, что есть значимая связь между зависимой и независимыми переменными.

AIC (Akaike Information Criterion) и BIC (Bayesian Information Criterion) - это меры сложности модели. Модели с более низкими значениями AIC и BIC считаются более предпочтительными.

Критический показатель Durbin-Watson (DW) используется для проверки наличия автокорреляции остатков. В данном случае DW равен 1,885, что означает, что автокорреляция отсутствует.
