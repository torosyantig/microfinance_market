```python
# создаем график
fig, ax = plt.subplots(figsize=(12, 6))

# добавляем линию инфляции
ax.plot(inflation['Дата'], inflation['Ср.знач ключевая ставки за квартал, % годовых'], color='blue', label='Ср.знач ключевая ставки за квартал, % годовых')

# добавляем точки инфляции
ax.scatter(inflation['Дата'], inflation['Ср.знач ключевая ставки за квартал, % годовых'], color='blue')

# добавляем линию ключевой ставки
ax.plot(inflation['Дата'], inflation['Ср. знач. инфляция за квартал, % г/г'], color='red', label='Ср. знач. инфляция за квартал, % г/г')

# добавляем точки ключевой ставки
ax.scatter(inflation['Дата'], inflation['Ср. знач. инфляция за квартал, % г/г'], color='red')

# добавляем заголовок
ax.set_title('Динамика инфляции и ключевой ставки')

# добавляем подписи к осям
ax.set_xlabel('Дата')
ax.set_ylabel('Индекс')

# добавляем легенду
ax.legend()

# добавляем горизонтальную линию на уровне 4%
ax.axhline(y=0.04, color='black', linestyle='--', linewidth=1)

# устанавливаем ограничения по оси y
ax.set_ylim([0, 0.25])
# получаем максимальные значения для инфляции и ключевой ставки
max_inf = inflation['Ср. знач. инфляция за квартал, % г/г'].max()
max_rate = inflation['Ср.знач ключевая ставки за квартал, % годовых'].max()

# добавляем подписи на график
ax.annotate(f'Max inflation: {max_inf:.3f}%', 
            xy=(inflation['Дата'][inflation['Ср. знач. инфляция за квартал, % г/г'].idxmax()], max_inf),
            xytext=(5, -10), textcoords='offset points', fontsize=12)
ax.annotate(f'Max key rate: {max_rate:.3f}%', 
            xy=(inflation['Дата'][inflation['Ср.знач ключевая ставки за квартал, % годовых'].idxmax()], max_rate),
            xytext=(5, -10), textcoords='offset points', fontsize=12)

# увеличиваем размер подписей на осях
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

# отображаем график
plt.show()
```
