```python
cons_conf = pd.read_excel('/Users/tigran/Projects/microfinance_market/data/Индекс_потребительской_уверенности_итог.xlsx', engine='openpyxl', header=0)
cons_conf['Дата']=pd.to_datetime(cons_conf['Дата'], errors='coerce')
cons_conf['Индекс уверенности'] = cons_conf['Индекс уверенности'].str.replace(',', '.').str.rstrip('%').astype(float) / 100
cons_conf['Индекс уверенности (в воозрасте до 30 лет)'] = cons_conf['Индекс уверенности (в воозрасте до 30 лет)'].str.replace(',', '.').str.rstrip('%').astype(float) / 100
cons_conf['Индекс уверенности (в воозрасте 30-50 лет) '] = cons_conf['Индекс уверенности (в воозрасте 30-50 лет)'].str.replace(',', '.').str.rstrip('%').astype(float) / 100
cons_conf['Индекс уверенности (в воозрасте от 50 лет)'] = cons_conf['Индекс уверенности (в воозрасте от 50 лет)'].str.replace(',', '.').str.rstrip('%').astype(float) / 100

cons_conf=cons_conf.drop('Индекс уверенности (в воозрасте 30-50 лет)', axis=1)
cons_conf=cons_conf.reindex(columns = ['Дата','Индекс уверенности', 'Индекс уверенности (в воозрасте до 30 лет)','Индекс уверенности (в воозрасте 30-50 лет)', 'Индекс уверенности (в воозрасте от 50 лет)'])
cons_conf.dtypes
```
